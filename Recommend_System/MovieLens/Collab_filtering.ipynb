```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv("movies.csv")
ratings = pd.read_csv("ratings.csv")
print(movies.shape)
print(ratings.shape)
```

```python
# Ratings matrix trasnform
ratings = ratings.drop("timestamp", axis=1)
ratings_matrix = ratings.pivot_table('rating', index='userId', columns='movieId')
ratings_matrix.head(3)
```

```python
rating_movies = pd.merge(ratings, movies, on="movieId")
rating_movies.head(3)
```

```python
# This time column and row transpose
movie_ratings_matrix = rating_movies.pivot_table('rating', index='title', columns='userId')
```

```python
movie_ratings_matrix = movie_ratings_matrix.fillna(0)
movie_ratings_matrix.head(3)
```

```python
# Cosine similarity
from sklearn.metrics.pairwise import cosine_similarity

item_based_sim = cosine_similarity(movie_ratings_matrix, movie_ratings_matrix)
item_based_sim
```

```python
# Result to DataFrame
item_based_sim_df = pd.DataFrame(data=item_based_sim, index=movie_ratings_matrix.index, 
                                 columns=movie_ratings_matrix.index)

item_based_sim_df.head(3)
```

```python
def get_item_based_collab(title):
    return item_based_sim_df[title].sort_values(ascending=False)[1:6]
```

```python
# Item Based Nearest Neighborhood CF
```

```python
def predict_rating(ratings_array, item_sim_array):
    ratings_pred = ratings_array.T.dot(item_sim_array) / np.array([np.abs(item_sim_array).sum(axis=1)])
    return ratings_pred
```

```python
ratings_pred = predict_rating(movie_ratings_matrix.values, item_based_sim_df.values) # 610x9719 dot 9719x9710
ratings_pred.shape
```

```python
rating_pred_matrix = pd.DataFrame(data=ratings_pred, index=movie_ratings_matrix.T.index, 
                                 columns = movie_ratings_matrix.T.columns)
rating_pred_matrix.head(3)
```

Predicted ratings could be smaller than actual ratings since the result of dot prodcut were divided by sum of consine similarity.

Actual raitngs 0 values are fiiled in predicted ratings values 

```python
# Compare acutal ratings and predicted ratings using MSE (only for non zero rating movies from actual)
from sklearn.metrics import mean_squared_error

def get_mse(actual, pred):
    actual_flat = actual[actual.nonzero()].flatten()
    pred_flat = pred[actual.nonzero()].flatten()
    
    return mean_squared_error(actual_flat, pred_flat)
```

```python
get_mse(movie_ratings_matrix.T.values, ratings_pred)
```

```python
# Top 20
def predict_rating_topn(ratings_array, item_sim_array, n=20):
    pred = np.zeros(ratings_array.T.shape)
    
    for col in range(ratings_array.T.shape[1]):
        top_n_items = [np.argsort(item_sim_array[:, col])[:-n-1:-1]]
        
        for row in range(ratings_array.T.shape[0]):
            pred[row, col] = item_sim_array[col, :][top_n_items].dot(ratings_array.T[row, :][top_n_items].T)
            pred[row, col] /= np.sum(np.abs(item_sim_array[col, :][top_n_items]))
        return pred
```

```python
rating_pred = predict_rating_topn(movie_ratings_matrix.values, item_based_sim_df.values, n=20)
```

```python
get_mse(movie_ratings_matrix.T.values, ratings_pred)
```

```python
rating_pred_matrix = pd.DataFrame(data=ratings_pred, index=movie_ratings_matrix.T.index, 
                                 columns = movie_ratings_matrix.T.columns)
rating_pred_matrix.head(3)
```

```python
user_rating_id = movie_ratings_matrix.T.loc[9, :]
user_rating_id[user_rating_id > 0].sort_values(ascending=False)[:10]
```

```python
def get_unseen_movie(movie_ratings_matrix, userId):
    user_rating = movie_ratings_matrix.T.loc[userId, :] #series
    
    already_seen = user_rating[user_rating_id > 0].index.tolist()
    movie_list = movie_ratings_matrix.T.columns.tolist()
    
    unseen_list = [movie for movie in movie_list if movie not in already_seen]
    
    return unseen_list
```

```python
def recom_movie_by_userid(pred_df, userId, unseen_list, top_n=10):
    recom_movies = pred_df.loc[userId, unseen_list].sort_values(ascending=False)[:top_n]
    return recom_movies
```

```python
unseen_list = get_unseen_movie(movie_ratings_matrix, 1)
```

```python
recom_movies = recom_movie_by_userid(rating_pred_matrix, 1, unseen_list, top_n=10)
```
