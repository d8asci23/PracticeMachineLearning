```python
import pandas as pd
import numpy as np
import warnings; warnings.filterwarnings('ignore')

movies = pd.read_csv("tmdb_5000_movies.csv")
print(movies.shape)
movies.head(2)
```

```python
movies_df = movies[['id', 'title', 'genres', 'vote_average', 'vote_count', 'popularity', 'keywords', 'overview']]
```

```python
movies_df.head(3)
```

```python
# weighted rating = (v/(v+m))*R + (m/(v+m))*C from imdb weighted rating calculation
# R = avg(rating) for the movie
# v = number of votes for the movie
# m = min votes required to be listed in the top 250
# c = the mean vote across the whole report

C = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.95)
print(C, " ", m)
#movies_df = movies_df.loc[movies_df['vote_count']>= m]
#print(movies_df.shape)


```

```python
def weighted_vote_average(data):
    v = data['vote_count']
    R = data['vote_average']
    
    return ((v/(v+m))*R) + ((m/(v+m))*C)

movies_df['weighted_vote'] = movies.apply(weighted_vote_average, axis=1)
```

```python
movies_df.head(3)
```

```python
pd.set_option('max_colwidth', 100)
movies_df[['genres', 'keywords']][:3]
```

```python
from ast import literal_eval
movies_df['genres'] = movies_df['genres'].apply(literal_eval)
movies_df['keywords'] = movies_df['keywords'].apply(literal_eval)
```

```python
movies_df[['genres', 'keywords']][:3]
```

```python
movies_df['genres'] = movies_df['genres'].apply(lambda x : [y['name'] for y in x])
movies_df['keywords'] = movies_df['keywords'].apply(lambda x : [y['name'] for y in x])
movies_df[['genres', 'keywords']].head(2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x : (' ').join(x))
count_vect = CountVectorizer(ngram_range=(1, 2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
print(genre_mat.shape)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:1])
```

```python
# find the indexes with high similarites
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
print(genre_sim_sorted_ind[:2])
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie_index = df[df['title'] == title_name].index.values
    similar_indexes = sorted_ind[title_movie_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
    
    similar_indexes = similar_indexes[similar_indexes != title_movie_index]

    
    return df.iloc[similar_indexes].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
similar_movies = find_sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average', 'weighted_vote']]
```
